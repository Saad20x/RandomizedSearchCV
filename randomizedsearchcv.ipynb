```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('../input/social-network-ads/Social_Network_Ads.csv')
```

```python
dataset
```

```python
X = dataset[['Age', 'EstimatedSalary']]
y = dataset[['Purchased']]
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=2 )
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train) 
```

```python
X_test = sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 50)
```

```python
rf.fit(X_train, y_train)
```

```python
from sklearn.model_selection import RandomizedSearchCV

from scipy.stats import randint
```

```python
est = RandomForestClassifier(n_jobs=-1)
rf_p_dist={'max_depth':[3,5,10,None],
              'n_estimators':[10,100,200,300,400,500],
              'max_features':randint(1,3),
               'criterion':['gini','entropy'],
               'bootstrap':[True,False],
               'min_samples_leaf':randint(1,4),
              }
```

```python
def hypertuning_rscv(est, p_distr, nbr_iter,X,y):
    rdmsearch = RandomizedSearchCV(est, param_distributions=p_distr,
                                  n_jobs=-1, n_iter=nbr_iter, cv=9)
    #CV = Cross-Validation ( here using Stratified KFold CV)
    rdmsearch.fit(X,y)
    ht_params = rdmsearch.best_params_
    ht_score = rdmsearch.best_score_
    return ht_params, ht_score

rf_parameters, rf_ht_score = hypertuning_rscv(est, rf_p_dist, 40, X, y)
```

```python
clf=RandomForestClassifier(n_jobs=-1, n_estimators=300,bootstrap= True,criterion='entropy',max_depth=3,max_features=2,min_samples_leaf= 3)
```

```python
y_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score=accuracy_score(y_test,y_pred)
accuracy_score
```

```python
from sklearn.model_selection import cross_val_score

cross_val=cross_val_score(clf,X,y,cv=10,scoring='accuracy').mean()
cross_val
```
